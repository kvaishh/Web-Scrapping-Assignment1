# Ques-1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location.

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\Asus\Desktop\New folder\chromedriver.exe")
```

```python
#opening the naukri page on automated chrome browser
driver.get("https://www.naukri.com/")
```

```python
designation=driver.find_element(By.CLASS_NAME,"suggestor-input")
designation.send_keys('Data Analyst')
```

```python
location=driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
location.send_keys('Banglore')
```

```python
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
#Scraping job title,job location,company name and experience required resp.
title_tags=driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)

location_tags=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi location"]')
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)

company_tags=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)
    
experience_tags=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi experience"]')
for i in experience_tags[0:10]:
    exp=i.text
    experience_required.append(exp)
                                   
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
#creating the dataframe from the above data 
df=pd.DataFrame({'Job_title':job_title,'Job_location':job_location,'Company_name':company_name,'Exp':experience_required})
df
```

```python
Rating_list=[]
Review_sum=[]
Full_review=[]
```

# Ques-10  Write a python program to display list of 50 Most expensive cars in the world (i.e.Car name ,Description and Price) from https://www.motor1.com/

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\Asus\Desktop\New folder\chromedriver.exe")
```

```python
#opening the page on automated chrome browser
driver.get("https://www.motor1.com/")
```

```python
#For clicking the hamburger menu
h_menu=driver.find_element(By.XPATH,'//*[@id="page_index_index_index"]/div[2]/div/div/div[1]/div')
h_menu.click()
```

```python
#For clicking the features button
dropdown=driver.find_element(By.XPATH,'/html/body/div[4]/div[1]/div[3]/ul/li[5]/button')
dropdown.click()
```

```python
#For clicking the lists option
lists=driver.find_element(By.XPATH,'/html/body/div[4]/div[1]/div[3]/ul/li[6]/ul/li[1]/a')
lists.click()
```

```python
#For clicking on the required link 50 most expensive cars
link=driver.find_element(By.XPATH,'//*[@id="page_index_articles_browse"]/div[9]/div[1]/div[1]/div/div/div[2]/div/div[1]/h3/a')
link.click()
```

```python
Car_Name=[]
Description=[]
Price=[]
```

```python
#Scraping car name,prices and description resp.
Name=driver.find_elements(By.CLASS_NAME,'subheader')
for i in Name:
    Car_Name.append(i.text)
Desc=driver.find_elements(By.TAG_NAME,'p')
for i in Desc:
    Description.append(i.text)
Prices=driver.find_elements(By.TAG_NAME,'strong')
for i in Prices:
    Price.append(i.text)
```

```python
df=pd.DataFrame({"Car_Name":Car_Name[0:50],"Prices":Prices[0:50],"Description":Description[0:50]})
df
```
